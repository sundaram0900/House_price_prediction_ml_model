#Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
boston = load_boston().data
df = pd.DataFrame(boston, columns=load_boston().feature_names)
df['MEDV'] = load_boston().target
```

#Making the dataset into dataframe

```python


data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
df = pd.DataFrame(data, columns=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT'])
df['PRICE'] = target
```

#Machine Learning - Linear regression

```python
df.head()
```

```python
#should always convert the data into array gorm to fit in the model
x=np.array(df.drop(['PRICE'],axis=1))
y=np.array(df['PRICE'])
```

```python
x
```

```python
y
```

# Splitting the data

```python
#spliting the data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

# choosing the model

```python
model = LinearRegression()
```

#Fitting/Train the model

```python
model.fit(x_train,y_train)
```

# prediction

```python
y_test
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_test,y_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('y_test')
plt.ylabel('y_pred')
plt.title('Actual vs predicted')
plt.grid()
plt.plot([min(y_test),max(y_test)],[min(y_pred),max(y_pred)],color='red')
plt.show()
```
